## Question-1. What is a Decision Tree, and how does it work?

* A Decision Tree is a popular machine learning algorithm used for both classification and regression tasks. It is a tree-like model that makes decisions based on a series of questions about the input features. Each internal node of the tree represents a feature (or attribute), each branch represents a decision rule, and each leaf node represents an outcome (or class label).

### How Decision Trees Work
#### Structure: 
* A Decision Tree consists of nodes and branches. The top node is called the root, and it represents the entire dataset. As you move down the tree, the dataset is split into subsets based on feature values.

### Splitting Criteria: 
* The algorithm selects the best feature to split the data at each node. Common criteria for determining the best split include:

#### Gini Impurity: 
* Measures the impurity of a dataset. A lower Gini score indicates a better split.
#### Entropy: 
* Measures the amount of disorder or uncertainty. The goal is to minimize entropy after the split.
#### Mean Squared Error (MSE): 
* Used for regression tasks to minimize the variance in the target variable.
### Recursive Partitioning: 
* The process of splitting continues recursively for each subset until a stopping criterion is met. This could be a maximum depth of the tree, a minimum number of samples in a node, or when no further information gain can be achieved.

### Leaf Nodes: 
* Once the tree reaches a stopping point, the final nodes (leaf nodes) represent the predicted output. For classification tasks, the leaf node will contain the majority class of the samples that reach that node. For regression tasks, it will contain the average value of the target variable.

### Prediction: 
* To make predictions, a new instance is passed down the tree, following the decision rules at each node until it reaches a leaf node, which provides the predicted outcome.

Advantages and Disadvantages
### Advantages:

* Easy to interpret and visualize.
* Requires little data preprocessing (no need for feature scaling).
* Can handle both numerical and categorical data.
### Disadvantages:

* Prone to overfitting, especially with deep trees.
* Sensitive to noisy data and outliers.
* Can be biased towards features with more levels.

## Question-2. What are impurity measures in Decision Trees?

* Impurity measures are essential for determining how to split data at each node in a Decision Tree. They quantify the level of disorder or uncertainty in a dataset, guiding the algorithm to make optimal splits.

### Gini Impurity:

* Definition: Gini impurity assesses the likelihood of misclassifying a randomly chosen element from the dataset.
* Formula: $$[ Gini(D) = 1 - \sum_{i=1}^{C} y_i^2 ]$$
### Entropy:

* Definition: Entropy measures the amount of uncertainty or disorder in a dataset, derived from information theory.
* Formula: $$[ Entropy(D) = - \sum_{i=1}^{C} y_i \log_2(p_i) ]$$
### Mean Squared Error (MSE):

* Definition: MSE is used for regression tasks, measuring the average squared difference between predicted and actual values.
* Formula: $$[ MSE(D) = \frac{1}{N} \sum_{i=1}^{N} (y_i - \bar{y})^2 ]$$
### Classification Error:

* Definition: This measure indicates the proportion of misclassified instances in a dataset.
* Formula: $$[ Error(D) = 1 - \max(p_i) ]$$ 

## Question-3.What is the mathematical formula for Gini Impurity?

* The Gini Impurity is a measure used to evaluate the impurity or purity of a dataset in the context of classification tasks. It quantifies the likelihood of misclassifying a randomly chosen element from the dataset. The mathematical formula for Gini Impurity is 

For a dataset ( D ) with ( C ) classes, the Gini Impurity ( Gini(D) ) is calculated using the formula:

* $$[ Gini(D) = 1 - \sum_{i=1}^{C} y_i^2 ]$$


### Example:
For a dataset with three classes (A, B, and C) with proportions 
$$( p_A = 0.5 )$$
$$( p_B = 0.3 )$$
$$( p_C = 0.2 )$$

$$ Gini(D) = 1 - (p_A^2 + p_B^2 + p_C^2) = 1 - (0.5^2 + 0.3^2 + 0.2^2) = 1 - (0.25 + 0.09 + 0.04) = 1 - 0.38 = 0.62 $$

This Gini Impurity value of 0.62 indicates a moderate level of impurity in the dataset. A Gini Impurity of 0 would indicate perfect purity (all instances belong to a single class).

## Question-4.What is the mathematical formula for Entropy?

* The Entropy is a measure used to evaluate the uncertainty or disorder in a dataset, derived from information theory. It quantifies the amount of uncertainty or disorder in a dataset. The mathematical formula for Entropy is

For a dataset ( D ) with ( C ) classes, the Entropy ( Entropy(D) ) is calculated using the formula:

* $$[ Entropy(D) = - \sum_{i=1}^{C} y_i \log_2(y_i) ]$$

### Example:

For a dataset with three classes (A, B, and C) with proportions 
$$( p_A = 0.5 )$$
$$( p_B = 0.3 )$$
$$( p_C = 0.2 )$$

* Calculate the individual contributions:

$$( p_A \log_2(p_A) = 0.5 \log_2(0.5) = 0.5 \times (-1) = -0.5 )$$
$$( p_B \log_2(p_B) = 0.3 \log_2(0.3) \approx 0.3 \times (-1.737) \approx -0.521 )$$
$$( p_C \log_2(p_C) = 0.2 \log_2(0.2) \approx 0.2 \times (-2.321) \approx -0.464 )$$

Sum the contributions: 
$$[ H(D) = -(-0.5 - 0.521 - 0.464) = 0.5 + 0.521 + 0.464 \approx 1.485 ]$$

This Entropy value of approximately 1.485 indicates a moderate level of uncertainty in the dataset

## Question-5.What is Information Gain, and how is it used in Decision Trees?

* Information Gain is a measure used to evaluate the improvement in the purity of the dataset after a split. It quantifies the difference in impurity between the parent and child nodes. Information Gain is calculated as the difference between the parent node's Gini Impurity and the weighted average of the child node's Gini Impurities.

### Calculate the Entropy of the Original Dataset: 
The first step is to compute the entropy of the entire dataset ( D ) before any splits. This is done using the entropy formula:

$$[ H(D) = - \sum_{i=1}^{C} p_i \log_2(p_i) ]$$


### Calculate the Entropy After the Split: 
For a feature ( A ) that can take on ( V ) different values, the dataset is split into ( V ) subsets ( D_v ) based on the values of ( A ). The entropy for each subset is calculated, and the weighted average of these entropies is computed:

$$[ H(D|A) = \sum_{v=1}^{V} \frac{|D_v|}{|D|} H(D_v) ]$$

### Calculate Information Gain: 
The Information Gain ( IG(D, A) ) for feature ( A ) is then calculated as the difference between the entropy of the original dataset and the weighted entropy after the split:

$$[ IG(D, A) = H(D) - H(D|A) ]$$

### Usage in Decision Trees
#### Feature Selection: 
During the construction of a Decision Tree, the algorithm evaluates all available features and calculates the Information Gain for each. The feature with the highest Information Gain is selected for the split at that node. This process is repeated recursively for each subset created by the split.

#### Building the Tree: 
By consistently choosing features that maximize Information Gain, the Decision Tree algorithm effectively reduces uncertainty and improves the purity of the resulting nodes. This leads to a more accurate model.

#### Stopping Criteria: 
Information Gain can also be used as a criterion for stopping the growth of the tree. If the Information Gain for a potential split is below a certain threshold, the algorithm may stop splitting further, indicating that the node is sufficiently pure or that further splits do not provide significant benefits.

## Question-7. What is the mathematical explanation behind Decision Trees?


#### 1. Structure of a Decision Tree
* Root Node: Represents the entire dataset.
* Internal Nodes: Represent features used for splitting the data.
* Branches: Indicate the outcome of a decision (result of a feature test).
* Leaf Nodes: Terminal nodes that represent the predicted outcome or class label.
#### 2. Splitting Criteria
* To determine the best feature for splitting the data, two common criteria are used: Gini Impurity and Entropy.

* Gini Impurity: 
$$[ Gini(D) = 1 - \sum_{i=1}^{C} p_i^2 ]$$ 


* Entropy: 
$$[ H(D) = - \sum_{i=1}^{C} p_i \log_2(p_i) ]$$ 

#### 3. Information Gain
Information Gain (IG) is used to evaluate the effectiveness of a feature in classifying the data. It is defined as the reduction in entropy after a dataset is split on a feature ( A ):

* $$[ IG(D, A) = H(D) - H(D|A) ]$$

* $$[ H(D|A) = \sum_{v=1}^{V} \frac{|D_v|}{|D|} H(D_v) ]$$


#### 4. Recursive Partitioning
* Start with the entire dataset at the root node.
* Calculate Information Gain (or Gini Impurity) for each feature.
* Choose the feature with the highest Information Gain to split the dataset.
* Create child nodes for each possible value of the chosen feature.
Repeat until a stopping criterion is met (e.g., maximum depth or pure node).
#### 5. Prediction
* Start at the root node and evaluate the feature corresponding to that node.
* Follow the branch corresponding to the feature's value for the instance.
* Continue until a leaf node is reached.
* The prediction is the class label (for classification) or the average value (for regression) at the leaf node.

## Question-10. What is the difference between Pre-Pruning and Post-Pruning	?

* Pre-pruning and post-pruning are both techniques used to reduce the complexity of decision trees. However, they differ in their implementation and choice of stopping criterion.

#### 1. Timing of Pruning
##### *Pre-Pruning:* 
* This technique occurs during the construction of the decision tree. The growth of the tree is halted based on certain criteria before it reaches its maximum depth or perfectly classifies the training data.

##### *Post-Pruning:* 
* This technique occurs after the decision tree has been fully constructed. The tree is first grown to its maximum size, and then certain nodes or subtrees are removed based on their performance on a validation set.

#### 2. Methodology
##### *Pre-Pruning:* 
* Involves setting specific criteria that must be met for a node to be split further.
* Maximum depth of the tree
* Minimum number of samples required to split a node
* Minimum number of samples required in a leaf node
* Minimum impurity decrease required for a split
##### *Post-Pruning:* 
* Involves evaluating the performance of the fully grown tree and removing nodes or subtrees that do not contribute significantly to the model's accuracy.

* Cost Complexity Pruning (CCP)
* Reduced Error Pruning
* Minimum Description Length (MDL)
#### 3. Complexity of the Model
##### *Pre-Pruning:* 
* By stopping the growth of the tree early, pre-pruning can lead to a simpler model from the outset. However, it may also result in underfitting if the stopping criteria are too strict.

##### *Post-Pruning:* 
* Allows for a more complex model initially, which can capture more intricate patterns in the data. The pruning process then simplifies the model as needed, potentially leading to a better balance between complexity and performance.

#### 4. Risk of Overfitting vs. Underfitting
##### *Pre-Pruning:* 
* Reduces the risk of overfitting by limiting the complexity of the model during construction. However, it may increase the risk of underfitting if the criteria are too conservative.

##### *Post-Pruning:* 
* Primarily addresses overfitting by simplifying a complex model after it has been built. It allows for a more flexible approach, as the model can initially capture more information before being pruned.

#### 5. Implementation
##### *Pre-Pruning:* 
* Often easier to implement since it involves setting parameters before the tree is built. It can be integrated into the tree-building algorithm itself.

##### *Post-Pruning:* 
* Requires a separate evaluation phase after the tree is built, which may involve additional computational overhead and the need for a validation dataset.



## Question-11. What is a Decision Tree Regressor?

* A Decision Tree Regressor is a type of decision tree that is used for regression tasks. It works by recursively partitioning the input space into smaller regions based on the values of features, and then predicting the target variable by averaging the values in each region.

* #### Tree Structure: 
The model is organized as a series of nodes and branches. Internal nodes represent decisions based on feature values, while leaf nodes represent the predicted output.

* #### Splitting Criteria: 
The decision tree regressor uses criteria such as Mean Squared Error (MSE) or Mean Absolute Error (MAE) to determine the best way to split the data at each node. The goal is to minimize the error in predictions.

* #### Recursive Partitioning: 
The algorithm recursively divides the dataset into subsets based on the feature that results in the best split. This process continues until a stopping condition is met, such as reaching a maximum depth or having a minimum number of samples in a leaf node.

* #### Interpretability: 
One of the main advantages of decision trees is their interpretability. The structure allows users to easily understand how decisions are made and which features are most influential in predicting the target variable.

* #### Handling Non-linearity: 
Decision tree regressors can capture non-linear relationships between features and the target variable without requiring any transformation of the input data.

## Question-13.How does a Decision Tree handle missing values?

* Decision trees handle missing values by using techniques such as imputation or deletion.

* #### Imputation: 
Imputation methods can be used to fill in missing values based on the available data. Common imputation methods include mean imputation, median imputation, mode imputation, or regression imputation.

* #### Deletion:
Missing values can be deleted from the dataset if they are not important for the analysis or if the number of missing values is small. This can be a good alternative to imputation, but it may lead to loss of important information.



## Question-14. How does a Decision Tree handle categorical features?

* Decision trees handle categorical features by using techniques such as one-hot encoding or ordinal encoding.

* #### One-hot Encoding: 
One-hot encoding creates a binary column for each unique value in the categorical feature. For example, if a categorical feature has three unique values (A, B, and C), the feature will be encoded as A_0, A_1, A_2, B_0, B_1, B_2, and C_0, C_1, C_2.

* #### Ordinal Encoding:
 Ordinal encoding assigns a unique integer value to each unique value in the categorical feature, without creating binary columns. For example, if a categorical feature has three unique values (A, B, and C), the feature will be encoded as 1, 2, 3. This encoding can be useful for algorithms that require numerical input, but it may not be as interpretable as one-hot encoding.

## *Question-27. Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and visualize its effect on accuracy*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

data = load_iris()
X = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(random_state=42)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(f'Initial Accuracy (without pruning): {accuracy_score(y_test, y_pred):.2f}')
```

```python
path = classifier.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
```

```python
accuracies = []

for alpha in ccp_alphas:
    dt_pruned = DecisionTreeClassifier(random_state=1, ccp_alpha=alpha)
    dt_pruned.fit(X_train, y_train)
    y_pred_pruned = dt_pruned.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred_pruned))
    print(accuracies)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, accuracies, marker='o', color='blue')
plt.title('Effect of Cost Complexity Pruning on Accuracy')
plt.xlabel('CCP Alpha')
plt.ylabel('Accuracy')
plt.xticks(ccp_alphas)
plt.grid()
plt.axhline(y=accuracy_score(y_test, y_pred), color='r', linestyle='--', label='Initial Accuracy')
plt.legend()
plt.show()
```

## *Question-30.Write a Python program to train a Decision Tree Classifier and use GridSearchCV to find the optimal values for max_depth and min_samples_split.*

```python
from sklearn.datasets import load_breast_cancer

breast_cancer = load_breast_cancer()
X = breast_cancer.data
y = breast_cancer.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
clf = DecisionTreeClassifier()
```

```python
param_grid = {
    'max_depth': [None, 2, 3, 4, 5, 6, 7, 8, 9, 10],
    'min_samples_split': [2, 3, 4, 5, 6, 7, 8, 9, 10]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, scoring='accuracy', cv=5, n_jobs=-1, verbose=1)
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
print(f'Best Parameters: {best_params}')
```

```python
best_classifier = DecisionTreeClassifier(**best_params, random_state=1)
best_classifier.fit(X_train, y_train)
```

```python
y_pred = best_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(f'Accuracy of Decision Tree Classifier with best parameters: {accuracy * 100:.2f}')
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=breast_cancer.target_names))
```
